28 movies in the TMDB data do not have any genre information, it's not appearing as missing values because there is a placeholder '[]' filling up the space. I will drop the rows with no genre information.

```python
#drop rows with missing genre using the index
df.drop(df[df['genres'] == '[]'].index, axis=0, inplace=True)
```

```python
df.shape
```

```python
def extract_genres():
    """Function to extract genre categories from the list of dictionaries 
    the genre information is currently stored in
    INPUT: Nil
    
    OUTPUT: List of list of genre categories
    """
    genre_list = []
    
    
    for genre in df['genres']:
        t = ast.literal_eval(re.search('({.+})', genre).group(0))
        
        try:
            l = [d['name'] for d in t if 'name' in d]
            genre_list.append(l)
        except:
            single_genres = []
            single_genres.append(t['name'])
            genre_list.append(single_genres)
    return genre_list
```

```python
def convert_list_to_string():
    """Convert a list of strings into one string, with the words separated 
    by a comma"""
    c = []
    for n in df['genre_list']:
        if len(n) == 1:
            c.append(n[0])
        else:
            joined_string = ",".join(n)
            c.append(joined_string)
    return c
```

```python
#the function needs the list of dictionaries to be a string to work
df['genres'] = df['genres'].astype(str)
```

```python
#call the function and save output in a new column in the dataframe
df['genre_list'] = extract_genres()
```

```python
#check for missing values
df.isnull().sum()
```

The genres are now in lists, next step would be to convert the list of strings into one string item

```python
df['cleaned_genres'] = convert_list_to_string()
```
