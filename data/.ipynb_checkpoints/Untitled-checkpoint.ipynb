```python
genre_key = {28:'Action', 12:'Adventure', 16:'Animation', 35:'Comedy', 80:'Crime',
             99:'Documentary', 18:'Drama', 10751:'Family', 14:'Fantasy', 36:'History', 
             27:'Horror', 10402:'Music', 9648:'Mystery', 10749:'Romance', 878 :'Science Fiction', 
             10770:'TV Movie', 53:'Thriller', 10752:'War',  37:'Western'}
```

```python
fetched_genres = []
for m in df['Title']:
    g = ''
    try:
        response = search.movie(query=m)
        response = response['results'][0]
        
        genre_ids = response['genre_ids']
        for k in genre_ids:
            if k in genre_key:
                g += genre_key[k] +', '
        fetched_genres.append(g)
    except:
        fetched_genres.append('cannot fetch')
```

```python
len(fetched_genres)
```

```python
df['fetched_genres'] = fetched_genres
```

```python
def extract_genres(df):
    gs = []
    for genre in df['genres']:
        try:
            t = ast.literal_eval(re.search('({.+})', genre).group(0))
        except:
            t = 'pass'
        try:
            l = [d['name'] for d in t if 'name' in d]
            gs.append(l)
        except:
            gs.append('not doing')
    return gs
```

```python
x = extract_genres(df)
```

```python
df['new'] = x
```

```python
c = []
for n in df['new']:
    joined_string = ",".join(n)
    c.append(joined_string)
```
